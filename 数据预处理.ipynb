```python
X[:, 2:]
```

```python
X[ : , 2:] = imputer.transform(X[ : , 2:]) # transform函数将数据标准化处理 一般对训练集数据，会先进行fit数据拟合再进行标准化处理
# 但是在测试集上，需要对数据进行标准化处理而不需要进行拟合
```
